### Use of arguments in functions *args

```python
def arguments(*args):
    for i in args:
        print(i)
```

```python
arguments(4, 5, 6, 7, 11, 13)
```

```python
def add_many(*nums):
    """
    This function adds multiple numbers
    """
    # Intialize sum to 0
    s = 0
    # Apply for loop on nums
    for i in nums:
        s = s + i
    return s       
```

```python
add_many(4,6)
```

```python
add_many(1,4,7,9)
```

```python
1+4+7+9
```

```python
s1 = add_many(10, 111, 13, 15, 17, 56, 82)
```

```python
s1
```

```python
10+111+13+15+17+56+82
```

### Getting product of multiple numbers

```python
def product_many(*nums):
    """
    This function gives product for multiple numbers
    """
    # Inialize product by 1
    p = 1

    # Apply for loop
    for i in nums:
        p = p*i

    return p
```

```python
product_many(3, 4)
```

```python
product_many(6, 10, 12, 14, 11)
```

```python
6*10*12*14*11
```

```python
def keyword_arguments(**kwargs):
    for key, value in kwargs.items():
        print(f'{key} : {value}')
```

### Function to Calclate hyptoenuse of right angled triangle
$ c^2 = a^2 + b^2 $

```python
def hypotenuse(a, b):
    c = (a**2 + b**2)**(1/2)
    return c
```

### Takes 3 sides of triangle as input and tell wheter it forms a triangle or not 
Sum of two sides > third

```python
def is_Triangle(x, y, z):
    """
    This function tells whether given 3 sides form a triangle or not
    """
    if (x+y)>z and (y+z)>x and (z+x)>y:
        s = True
    else:
        s = False
    return s
```

```python
is_Triangle(3,4,5)
```

```python
is_Triangle(1,2,4)
```

### Calculating area of triangle with given sides a, b, c
Herons formula
$ s = (a+b+c)/2 $

$ area = sqrt(s*(s-a)*(s-b)*(s-c)) $

```python
def area_triangle(a, b, c):
    if is_Triangle(a,b,c)==True:
        s = (a+b+c)/2
        ar = (s*(s-a)*(s-b)*(s-c))**(1/2)
    else:
        ar = None
        print(f'The given sides {a}, {b} and {c} DO NOT Form a triangle.')
    return ar
```

```python
t1 = area_triangle(12, 13, 14)
```

```python
t1
```

```python
t2 = area_triangle(1,2,4)
```

```python
t2
```

```python
area_triangle(3,4,5)
```
